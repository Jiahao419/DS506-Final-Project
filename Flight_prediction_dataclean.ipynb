```python
import pandas as pd
import glob
import os
import polars as pl
```

```python
base_path = "data"
csv_files = glob.glob(os.path.join(base_path, "**", "*.csv"), recursive=True)

output_file = "merged_flights.csv"
first = True

for f in csv_files:
    print(f)
    # read in chunks to avoid memory overload
    for chunk in pd.read_csv(f, chunksize=100000):
        chunk.to_csv(output_file, mode="a", index=False, header=first)
        first = False

print(f"Merged {len(csv_files)} CSVs into {output_file}")
```

```python
import pandas as pd

usecols = [
    "FlightDate",
    "IATA_Code_Operating_Airline",
    "Origin",
    "OriginCityName",
    "Dest",
    "DestCityName",
    "DepDelayMinutes",
    "DepartureDelayGroups",
    "ArrDelayMinutes",
    "ArrivalDelayGroups",
    "Cancelled",
    "CancellationCode",
    "Diverted",
    "CarrierDelay",
    "WeatherDelay",
    "NASDelay",
    "SecurityDelay",
    "LateAircraftDelay",
]

chunksize = 10**6
output_file = "flights_cleaned.csv"

first_chunk = True
for chunk in pd.read_csv("merged_flights.csv", usecols=usecols, chunksize=chunksize):
    chunk = chunk.dropna(subset=["DepDelayMinutes"])
    chunk.to_csv(
        output_file,
        mode="w" if first_chunk else "a",
        header=first_chunk,
        index=False
    )
    first_chunk = False
```
